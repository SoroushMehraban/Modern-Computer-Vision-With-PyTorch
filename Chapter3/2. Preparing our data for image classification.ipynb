```python
from torchvision import datasets
import matplotlib.pyplot as plt
import numpy as np
import torch
```

## Downloading the dataset

```python
data_folder = 'dataset/'
fmnist = datasets.FashionMNIST(data_folder,
                               download=True,
                               train=True  # Only download the training images
                              )
```

```python
tr_images = fmnist.data # images
tr_targets = fmnist.targets # labels
```

```python
unique_values = tr_targets.unique()
print(f'tr_images & tr_targets:\n\tX -{tr_images.shape}\n\tY \
-{tr_targets.shape}\n\tY-Unique Values : {unique_values}')
print(f'TASK:\n\t{len(unique_values)} class Classification')
print(f'UNIQUE CLASSES:\n\t{fmnist.classes}')
```

## Plot random sample of 10 images for all 10 possible classes

```python
number_of_rows, number_of_columns = len(tr_targets.unique()), 10
fig, ax = plt.subplots(number_of_rows, number_of_columns, figsize=(10, 10))
for label_class, plot_row in enumerate(ax):
    label_x_rows = np.where(tr_targets == label_class)[0]  # returns a tuple that first index is the desired array
    for plot_cell in plot_row:
        plot_cell.grid(False)
        plot_cell.axis('off')
        
        ix = np.random.choice(label_x_rows)
        
        image = tr_images[ix]
        plot_cell.imshow(image, cmap='gray')
```
