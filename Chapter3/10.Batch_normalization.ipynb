```python
from torchvision import datasets
from torch.utils.data import Dataset, DataLoader
from torch.optim import SGD, Adam
import matplotlib.pyplot as plt
import torch.nn as nn
import numpy as np
import torch
```

# Global variables

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
data_folder = "dataset/"
```

### Loading Training Data

```python
tr_fmnist = datasets.FashionMNIST(data_folder, download=True, train=True)
tr_images, tr_targets = tr_fmnist.data, tr_fmnist.targets
```

### Loading Test Data

```python
val_fmnist = datasets.FashionMNIST(data_folder, download=True, train=False)
val_images, val_targets = val_fmnist.data, val_fmnist.targets
```

# Functions And Classes

```python
class FMNISTDataset(Dataset):
    def __init__(self, x, y):
        x = x.float() / (255 * 10000)  # Done only for us to understand the impact of batch normalization
        x = x.view(-1, 28 * 28)
        self.x, self.y = x, y
    
    def __getitem__(self, ix):
        return self.x[ix].to(device), self.y[ix].to(device)
    
    def __len__(self):
        return self.x.shape[0]
```

```python
def get_model():
    class neuralnet(nn.Module):
        def __init__(self):
            super().__init__()
            self.input_to_hidden_layer = nn.Linear(784, 1000)
            self.hidden_layer_activation = nn.ReLU()
            self.hidden_to_output_layer = nn.Linear(1000, 10)
        def forward(self, x):
            x = self.input_to_hidden_layer(x)
            x1 = self.hidden_layer_activation(x)
            x2 = self.hidden_to_output_layer(x1)
            return x2, x1
    model = neuralnet().to(device)
    loss_function = nn.CrossEntropyLoss()
    optimizer = Adam(model.parameters(), lr=1e-3)
    return model, optimizer, loss_function
```

```python
def train_batch(x, y, model, optimizer, loss_function):
    model.train()
    prediction = model(x)[0]
    batch_loss = loss_function(prediction, y)
    batch_loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    return batch_loss.item()
```

```python
def accuracy(x, y, model):
    model.eval()
    with torch.no_grad():
        prediction = model(x)[0]
    max_values, argmaxes = prediction.max(-1)
    is_correct = argmaxes == y
    return is_correct.cpu().numpy().tolist()
```

```python
def get_data(batch_size):
    train_data = FMNISTDataset(tr_images, tr_targets)
    train_dataloader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    
    val_data = FMNISTDataset(val_images, val_targets)
    val_dataloader = DataLoader(val_data, batch_size=val_images.shape[0], shuffle=False)
    
    return train_dataloader, val_dataloader
```

```python
@torch.no_grad()
def val_loss(x, y, model, loss_function):
    model.eval()
    prediction = model(x)[0]
    val_loss = loss_function(prediction, y)
    return val_loss.item()
```

```python
train_dataloader, val_datalader = get_data(batch_size=32)
model, optimizer, loss_function = get_model()
```

# Training model with small range input vector

```python
train_losses, train_accuracies = [], []
val_losses, val_accuracies = [], []
for epoch in range(100):
    if epoch % 10 == 0:
        print(f"Current epoch: {epoch}")
    
    train_epoch_losses, train_epoch_accuracies = [], []
    
    for batch in train_dataloader:
        x, y = batch
        batch_loss = train_batch(x, y, model, optimizer, loss_function)
        train_epoch_losses.append(batch_loss)
    train_epoch_loss = np.mean(train_epoch_losses)
    
    for batch in train_dataloader:
        x, y = batch
        is_correct = accuracy(x, y, model)
        train_epoch_accuracies.extend(is_correct)
    train_epoch_accuracy = np.mean(train_epoch_accuracies)
    
    for batch in val_datalader:
        x, y = batch
        val_is_correct = accuracy(x, y, model)
        val_epoch_loss = val_loss(x, y, model, loss_function)
        
    val_epoch_accuracy = np.mean(val_is_correct)
    
    train_losses.append(train_epoch_loss)
    train_accuracies.append(train_epoch_accuracy)
    val_losses.append(val_epoch_loss)
    val_accuracies.append(val_epoch_accuracy)
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

epochs = [i for i in range(1, 101)]

# Add data to ax0
ax0.scatter(x=epochs,
            y=train_losses)
ax0.plot(val_losses, color='red')

# Customize ax0
ax0.set(title= "Training and validation loss with very small input values",
        ylabel= "Loss",
        xlabel= "Epochs")

# Add legend to ax0
ax0.legend(['Training loss', 'Validation loss'])

# Set Major ticks
ax0.xaxis.set_major_locator(plt.MultipleLocator(10))

# Configure the grid lines
ax0.grid('off')

# Add data to ax1
ax1.scatter(x=epochs,
            y=train_accuracies)
ax1.plot(val_accuracies, color='red')

# Customize ax1
ax1.set(title= "Training and validation accuracy with very small input values",
        ylabel= "Accuracy",
        xlabel= "Epochs")

# Add legend to ax1
ax1.legend(['Training accuracy', 'Validation accuracy'])

# Set Major ticks
ax1.xaxis.set_major_locator(plt.MultipleLocator(10))

# Configure the grid lines
ax1.grid('off')

# Add padding between axes
plt.tight_layout();
```

# With batch normalization

```python
def get_model():
    class neuralnet(nn.Module):
        def __init__(self):
            super().__init__()
            self.input_to_hidden_layer = nn.Linear(784, 1000)
            self.batch_norm = nn.BatchNorm1d(1000)
            self.hidden_activation_layer = nn.ReLU()
            self.hidden_to_output_layer = nn.Linear(1000, 10)
        def forward(self, x):
            x = self.input_to_hidden_layer(x)
            x0 = self.batch_norm(x)
            x1 = self.hidden_activation_layer(x0)
            x2 = self.hidden_to_output_layer(x1)
            return x2, x1
    
    model = neuralnet().to(device)
    loss_function = nn.CrossEntropyLoss()
    optimizer = Adam(model.parameters(), lr=1e-3)
    return model, optimizer, loss_function
```

```python
train_losses, train_accuracies = [], []
val_losses, val_accuracies = [], []
for epoch in range(100):
    if epoch % 10 == 0:
        print(f"Current epoch: {epoch}")
    
    train_epoch_losses, train_epoch_accuracies = [], []
    
    for batch in train_dataloader:
        x, y = batch
        batch_loss = train_batch(x, y, model, optimizer, loss_function)
        train_epoch_losses.append(batch_loss)
    train_epoch_loss = np.mean(train_epoch_losses)
    
    for batch in train_dataloader:
        x, y = batch
        is_correct = accuracy(x, y, model)
        train_epoch_accuracies.extend(is_correct)
    train_epoch_accuracy = np.mean(train_epoch_accuracies)
    
    for batch in val_datalader:
        x, y = batch
        val_is_correct = accuracy(x, y, model)
        val_epoch_loss = val_loss(x, y, model, loss_function)
        
    val_epoch_accuracy = np.mean(val_is_correct)
    
    train_losses.append(train_epoch_loss)
    train_accuracies.append(train_epoch_accuracy)
    val_losses.append(val_epoch_loss)
    val_accuracies.append(val_epoch_accuracy)
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

epochs = [i for i in range(1, 101)]

# Add data to ax0
ax0.scatter(x=epochs,
            y=train_losses)
ax0.plot(val_losses, color='red')

# Customize ax0
ax0.set(title= "Training and validation loss with very small input values and batch normalization",
        ylabel= "Loss",
        xlabel= "Epochs")

# Add legend to ax0
ax0.legend(['Training loss', 'Validation loss'])

# Set Major ticks
ax0.xaxis.set_major_locator(plt.MultipleLocator(10))

# Configure the grid lines
ax0.grid('off')

# Add data to ax1
ax1.scatter(x=epochs,
            y=train_accuracies)
ax1.plot(val_accuracies, color='red')

# Customize ax1
ax1.set(title= "Training and validation accuracy with very small input values and batch normalization",
        ylabel= "Accuracy",
        xlabel= "Epochs")

# Add legend to ax1
ax1.legend(['Training accuracy', 'Validation accuracy'])

# Set Major ticks
ax1.xaxis.set_major_locator(plt.MultipleLocator(10))

# Configure the grid lines
ax1.grid('off')

# Add padding between axes
plt.tight_layout();
```
