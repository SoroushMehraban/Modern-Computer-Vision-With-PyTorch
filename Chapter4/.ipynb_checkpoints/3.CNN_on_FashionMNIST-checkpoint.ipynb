```python
from torchvision import datasets
from torch.utils.data import Dataset, DataLoader
from torch.optim import SGD, Adam
import matplotlib.pyplot as plt
from torchsummary import summary
import torch.nn as nn
import numpy as np
import torch
```

# Global Variables

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
data_folder = 'dataset/'
```

## loading training data

```python
tr_fmnist = datasets.FashionMNIST(data_folder, download=True, train=True)
tr_images, tr_targets = tr_fmnist.data, tr_fmnist.targets
```

## loading validation data

```python
val_fmnist = datasets.FashionMNIST(data_folder, download=True, train=False)
val_images, val_targets = val_fmnist.data, val_fmnist.targets
```

# Functions and classes

```python
class FMNISTDataset(Dataset):
    def __init__(self, x, y):
        x = x.float() / 255
        x = x.view(-1, 1, 28, 28) # BatchSize×Channels×Height×Width
        self.x, self.y = x, y
    
    def __getitem__(self, ix):
        return self.x[ix].to(device), self.y[ix].to(device)
    
    def __len__(self):
        return self.x.shape[0]
```

```python
def get_model():
    model = nn.Sequential(
        nn.Conv2d(1, 64, kernel_size=3),
        nn.MaxPool2d(2),
        nn.ReLU(),
        nn.Conv2d(64, 128, kernel_size=3),
        nn.MaxPool2d(2),
        nn.ReLU(),
        nn.Flatten(),
        nn.Linear(3200, 256),
        nn.ReLU(),
        nn.Linear(256, 10)
    ).to(device)
#     class CNN(nn.Module):
#         def __init__(self):
#             super().__init__()
#             self.conv1 = nn.Conv2d(1, 64, kernel_size=3)
#             self.max_pool = nn.MaxPool2d(2)
#             self.relu = nn.ReLU()
#             self.conv2 = nn.Conv2d(64, 128, kernel_size=3)
#             self.flatten = nn.Flatten()
#             self.linear1 = nn.Linear(3200, 256)
#             self.linear2 = nn.Linear(256, 10)
#         def forward(self, x):
#             x = self.conv1(x)
#             x = self.max_pool(x)
#             x = self.relu(x)
#             x = self.conv2(x)
#             x = self.max_pool(x)
#             x = self.relu(x)
#             x = self.flatten(x)
#             x = self.linear1(x)
#             x = self.relu(x)
#             x = self.linear2(x)
#             return x
    
    # model = CNN().to(device)        
    
    loss_function = nn.CrossEntropyLoss()
    optimizer = Adam(model.parameters(), lr=1e-3)
    return model, loss_function, optimizer
```

```python
def train_batch(x, y, model, optimizer, loss_function):
    model.train()
    prediction = model(x)
    batch_loss = loss_function(prediction, y)
    batch_loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    return batch_loss.item()
```

```python
def accuracy(x, y, model):
    model.eval()
    with torch.no_grad():
        prediction = model(x)
    max_values, argmaxes = prediction.max(-1)
    is_correct = argmaxes == y
    return is_correct.cpu().numpy().tolist()
```

```python
def get_data(batch_size):
    train_data = FMNISTDataset(tr_images, tr_targets)
    train_dataloader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    
    val_data = FMNISTDataset(val_images, val_targets)
    val_dataloader = DataLoader(val_data, batch_size=val_images.shape[0], shuffle=False)
    
    return train_dataloader, val_dataloader
```

```python
@torch.no_grad()
def val_loss(x, y, model, loss_function):
    model.eval()
    prediction = model(x)
    val_loss = loss_function(prediction, y)
    return val_loss.item()
```

# Model Summary

```python
model, loss_function, optimizer = get_model()
# summary(model, torch.zeros(1, 1, 28, 28));
```

# Training the model

```python
train_dataloader, val_datalader = get_data(batch_size=32)
model, loss_function, optimizer = get_model()
```

```python
train_losses, train_accuracies = [], []
val_losses, val_accuracies = [], []
for epoch in range(5):
    print(f"Current epoch: {epoch}")
    
    train_epoch_losses, train_epoch_accuracies = [], []
    
    for batch in train_dataloader:
        x, y = batch
        batch_loss = train_batch(x, y, model, optimizer, loss_function)
        train_epoch_losses.append(batch_loss)
    train_epoch_loss = np.mean(train_epoch_losses)
    
    for batch in train_dataloader:
        x, y = batch
        is_correct = accuracy(x, y, model)
        train_epoch_accuracies.extend(is_correct)
    train_epoch_accuracy = np.mean(train_epoch_accuracies)
    
    for batch in val_datalader:
        x, y = batch
        val_is_correct = accuracy(x, y, model)
        val_epoch_loss = val_loss(x, y, model, loss_function)
        
    val_epoch_accuracy = np.mean(val_is_correct)
    
    train_losses.append(train_epoch_loss)
    train_accuracies.append(train_epoch_accuracy)
    val_losses.append(val_epoch_loss)
    val_accuracies.append(val_epoch_accuracy)
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

epochs = [i for i in range(1, 6)]

# Add data to ax0
ax0.scatter(x=epochs,
            y=train_losses)
ax0.plot(val_losses, color='red')

# Customize ax0
ax0.set(title= "Training and validation loss with CNN",
        ylabel= "Loss",
        xlabel= "Epochs")

# Add legend to ax0
ax0.legend(['Training loss', 'Validation loss'])


# Configure the grid lines
ax0.grid('off')

# Add data to ax1
ax1.scatter(x=epochs,
            y=train_accuracies)
ax1.plot(val_accuracies, color='red')

# Customize ax1
ax1.set(title= "Training and validation accuracy with CNN",
        ylabel= "Accuracy",
        xlabel= "Epochs")

# Add legend to ax1
ax1.legend(['Training accuracy', 'Validation accuracy'])


# Configure the grid lines
ax1.grid('off')

# Add padding between axes
plt.tight_layout();
```

# Effect of image translation

```python
preds = []
ix = 24300
for px in range(-5, 6):
    img = tr_images[ix] / 255.
    img = img.view(28, 28)
    img2 = np.roll(img, px, axis=1)
    plt.imshow(img2)
    plt.show()
    img3 = torch.Tensor(img2).view(-1, 1, 28, 28).to(device)
    np_output = model(img3).cpu().detach().numpy()
    preds.append(np.exp(np_output) / np.sum(np.exp(np_output)))
```

```python
import seaborn as sns
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
plt.title('Probability of each class for various translations')
sns.heatmap(np.array(preds).reshape(11, 10), annot=True, ax=ax, fmt='.2f', xticklabels=tr_fmnist.classes,
            yticklabels=[str(i)+str(' pixels') for i in range(-5,6)], cmap='gray');
```
