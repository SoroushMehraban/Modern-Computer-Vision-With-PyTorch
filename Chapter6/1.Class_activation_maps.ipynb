```python
from sklearn.model_selection import train_test_split
import numpy as np, cv2, pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch import optim

from torch.utils.data import Dataset, DataLoader
import torchvision
from torchvision import transforms as T, models, datasets

import os
from glob import glob
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
class_indices = {
    "Parasitized": 0,
    "Uninfected": 1
}
```

# Transformations

```python
train_transformations = T.Compose([
                                   T.ToPILImage(),
                                   T.Resize(128),  # minimum size of one dimension is 128
                                   T.CenterCrop(128),  # a square crop from center (output size: 128x128)
                                   T.ColorJitter(brightness=(0.95, 1.05),
                                                 contrast=(0.95, 1.05),
                                                 saturation=(0.95, 1.05),
                                                 hue=0.05  # hue factor is [-0.05, 0.05]
                                                 ),
                                   T.RandomAffine(degrees=5,  # (-5°, 5°) rotation
                                                  translate=(0.01,  # horizontal translation between -128 * 0.01 and 128 * 0.01
                                                             0.1  # vertical translation between -128 * 0.1 and 128 * 0.1
                                                             )
                                                  ),
                                   T.ToTensor(),  # (H, W, C) PIL Image range [0-255] -> (C, H, W) tensor range [0-1]
                                   T.Normalize(mean=[0.5, 0.5, 0.5],
                                               std=[0.5, 0.5, 0.5])  # (value - mean) / std
])
```

```python
val_transformations = T.Compose([
                                 T.ToPILImage(),
                                 T.Resize(128),
                                 T.CenterCrop(128), 
                                 T.ToTensor(),
                                 T.Normalize(mean=[0.5, 0.5, 0.5],
                                               std=[0.5, 0.5, 0.5])
])
```

# Dataset

```python
class MalariaImages(Dataset):
    def __init__(self, files, transform=None):
        self.files = files
        self.transform = transform
    
    def __len__(self):
        return len(self.files)
    
    def __getitem__(self, ix):
        file_path = self.files[ix]
        img_class = file_path.split("/")[1]

        img = cv2.imread(file_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        return img, img_class
    
    def collate_fn(self, batch):
        _images, classes = list(zip(*batch))
        if self.transform:
            images = [self.transform(img)[None] for img in _images]
        else:
            images = _images

        classes = [torch.tensor([class_indices[img_class] 
                                 for img_class in classes])]

        images, classes = [torch.cat(item).to(device) 
                          for item in [images, classes]]
        return images, classes, _images
```

```python
all_files = glob('cell_images/*/*.png')
np.random.seed(10)
np.random.shuffle(all_files)
```

```python
train_files, val_files =  train_test_split(all_files, random_state=1)

train_dataset = MalariaImages(train_files, transform=train_transformations)
val_dataset = MalariaImages(val_files, transform=val_transformations)

train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True,
                              collate_fn=train_dataset.collate_fn)
val_dataloader = DataLoader(val_dataset, batch_size=32, shuffle=False,
                              collate_fn=val_dataset.collate_fn)
```

# Model definition

```python
class MalariaClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            self.conv_block(3, 64),
            self.conv_block(64, 64),
            self.conv_block(64, 128),
            self.conv_block(128, 256),
            self.conv_block(256, 512),
            self.conv_block(512, 64),
            nn.Flatten(),
            nn.Linear(256, 256),
            nn.Dropout(0.2),
            nn.ReLU(inplace=True),
            nn.Linear(256, 2)
        )
        self.loss_function = nn.CrossEntropyLoss()
    
    def forward(self, x):
        return self.model(x)
    
    def compute_metrics(self, preds, targets):
        loss = self.loss_function(preds, targets)
        accuracy = (torch.max(preds, 1)[1] == targets).float().mean()
        return loss, accuracy

    @staticmethod
    def conv_block(in_channels, out_channels):
        return nn.Sequential(
            nn.Dropout(0.2),
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.BatchNorm2d(out_channels),
            nn.MaxPool2d(kernel_size=2),
        )
```

# Functions

```python
def train_batch(model, data, optimizer, criterion):
    model.train()
    images, labels, _ = data
    preds = model(images)
    loss, accuracy = criterion(preds, labels)

    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    return loss.item(), accuracy.item()
```

```python
@torch.no_grad()
def validate_batch(model, data, criterion):
    model.eval()
    images, labels, _ = data
    preds = model(images)
    loss, accuracy = criterion(preds, labels)
    return loss.item(), accuracy.item()
```

# Train the model

```python
model = MalariaClassifier().to(device)
criterion = model.compute_metrics
optimizer = optim.Adam(model.parameters(), lr=1e-3)
n_epochs = 2
```

```python
for epoch in range(1, n_epochs + 1):
    train_epoch_losses, train_epoch_accuracies = [], []
    for batch in train_dataloader:
        loss, accuracy = train_batch(model, batch, optimizer, criterion)
        train_epoch_losses.append(loss)
        train_epoch_accuracies.append(accuracy)
    
    val_epoch_losses, val_epoch_accuracies = [], []
    for batch in val_dataloader:
        loss, accuracy = validate_batch(model, batch, criterion)
        val_epoch_losses.append(loss)
        val_epoch_accuracies.append(accuracy)
    
    print(f"EPOCH: {epoch}\t"
          f"train_loss: {np.mean(train_epoch_losses):.2f}\t"
          f"train_accuracy: {np.mean(train_epoch_accuracies):.2f}\t"
          f"val_loss: {np.mean(val_epoch_losses):.2f}\t"
          f"val_accuracy: {np.mean(val_epoch_accuracies):.2f}")
```

# Produce the heatmap

![Grad-CAM](images/grad-cam.jpg)

**Steps:**
1. Pass the input image through the network and get the classification output.
2. Pick a class on the output layer that you want to make a heatmap out of it. e.g. Tiger Cat
3. Calculate backward gradients of the last conv layer with respect to the picked class. Now for each conv layer's channel you have their gradients.
4. Calculate Global Average Pooling (GAP) of each channel to get $w_{1}^c$, $w_{2}^c$, ..., $w_{n}^c$.
5. Multiply each $w_{i}^c$ with the corresponding feature map. Then take the average of them.
6. Pass through the ReLU (Not implemented in this code)

```python
# Remove the FCNNs at the end
img_to_feature_map = nn.Sequential(*(list(model.model[:5].children()) 
                                   + list(model.model[5][:2].children())))
```

```python
def im2gradCAM(model, img_to_feature_map, x):
    model.eval()    
    logits = model(x)
    activations = img_to_feature_map(x)
    
    # logits is 1x2, so .max(1) gives (maximum, maximum_index) and we get 
    # the index as the prediction
    pred = logits.max(1)[1] 
    
    model.zero_grad()

    # backward based on the prediction
    logits[0,pred].backward(retain_graph=True)  

    last_conv_layer = model.model[-6][1]
    # Global Average Pooling over different activations
    pooled_grads = last_conv_layer.weight.grad.data.mean((1,2,3))

    # multiply each activation map with corresponding gradient average
    for i in range(activations.shape[1]):
        activations[:,i,:,:] *= pooled_grads[i]
    
    # Producing heatmap layer by taking average across different activations
    heatmap = torch.mean(activations, dim=1)[0].cpu().detach()

    return heatmap, 'Uninfected' if pred.item() else 'Parasitized'
```

```python
def upsample_heatmap(heatmap, img, img_original_size=128):
    # convert entries range to [0, 255]
    heatmap = 255 * ((heatmap - heatmap.min()) / (heatmap.max() - heatmap.min()))
    
    heatmap = np.uint8(heatmap)
    heatmap = cv2.resize(heatmap, (img_original_size, img_original_size))
    heatmap = cv2.applyColorMap(255 - heatmap, cv2.COLORMAP_JET)
    heatmap = np.uint8(heatmap)
    
    # Combination of heatmap and img
    out_img = np.uint8(heatmap * 0.7 + img * 0.3) 
    
    return out_img
```

```python
img_numbers = 20
test_dataloader =  DataLoader(val_dataset,
                              batch_size=img_numbers,
                              shuffle=True,
                              collate_fn=val_dataset.collate_fn)

imgs, classes, orginal_imgs = next(iter(test_dataloader))
for i in range(img_numbers):
    image = cv2.resize(orginal_imgs[i], (128, 128))
    heatmap, prediction = im2gradCAM(model, img_to_feature_map,imgs[i:i+1])
    if prediction == 'Uninfected':
        continue

    heatmap = upsample_heatmap(heatmap, image)

    # plot the cell
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))

    ax[0].imshow(image)
    ax[0].set(title="Original image")

    ax[1].imshow(heatmap)
    ax[1].set(title="Heatmap")

    plt.show()
    print("-------------------------")
```
